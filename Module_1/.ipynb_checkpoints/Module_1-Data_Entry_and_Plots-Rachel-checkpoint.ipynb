## Entering Data: SFU Enrollment

We start by importing some data. This particular datset contains information about enrollment at SFU for the 2015-2016 academic year divided up by faculty. The two variables in this dataset are called "faculty" and "students". The "faculty" variable tells us the name of each faculty, and "students" tells us how many students are enrolled in each faculty. This is a categorical dataset. The individuals are students, and each student is from one of the faculties.

We start by entering these variables separately. The faculty names are abbreviated so that they will fit in our plots later.

```r
faculty.input = c("ASci", "FASS", "Bus", "CAT", "Educ", "Env", "HSci", "Sci", "Other")
students.input = c(3408, 4880, 3716, 2766, 899, 981, 1435, 2742, 70)
```

- ALSO, DO YOU PUT .input AT THE END OF THE VARIABLE NAMES IN ORDER TO AVOID CONFUSION BETWEEN THE INDIVIDUAL VARIABLES AND THOSE STORED AS PART OF THE data.fac DATA.FRAME?  
- DOES c STAND FOR CONCATENATE? IF SO, USEFUL TO SAY?

We enter the values of each variable inside the "c" function, and separate the values with commas. The "c" function is like glue that sticks the individual values together into a single variable.

Notice that the values of faculty are entered in quotation marks. This tells R that they are words (instead of names of other things in R). Anytime you want to enter words into R, you have to put them in quotations. 

Notice the extra ".input" at the end of the names of our vaiables. This is there because we still need to combine these variables into a single dataset, called a "data frame". We use the "data.frame" function to accomplish this.

```r
data.fac = data.frame(faculty = faculty.input, students = students.input)
```

We now have a data frame called "data.fac" that contains two variables, "faculty" and "students". We can print this data frame to make sure we input everything correctly (it's actually a good idea to print each of the intermediate steps as well to make sure nothing goes wrong there).

```r
print(faculty.input)
print(students.input)
print(data.fac)
```

Notice that the rows in our dataset are numbered. This is just for convenience, and won't actually interfere with any of our analysis.

We can get back either the faculty or students variable from our dataset using "$". For example, let's print just the number of students in each faculty.

```r
print(data.fac$students)
```

## Pie Charts

Now that we have entered our dataset, let's investigate its distribution using plots. First, let's make a pie chart. This is done in R using the "pie" function. 

The "pie" function uses two inputs separated by a comma. The first input is called "x", and it is the proportion or number of things in each category (in our case, the number of students in each faculty). The second input is called "labels", and it is the labels for each category (in our case, the name of each faculty). The category names can be omitted (which gives a pie chart without labels), but you usually need some way to tell the slices apart. Let's put this information together now and make a pie chart.

```r
pie(x = data.fac$students, labels = data.fac$faculty)
```

You can also add a title to your pie chart. This is done using another optional input to the "pie" function called "main". We want R to read this title as words, so we need to enclose it in quotations.

```r
pie(x = data.fac$students, labels = data.fac$faculty, main="Pie Chart of Student Enrollment by Faculty")
```

## Bar Charts

We now move on to bar charts. The R function for bar charts is barplot. This function has a three important inputs called "height", "names.arg" and "las". The "height" input tells R how tall to draw each bar. The "names.arg" input tells R how to label each bar. As with the pie chart, we can optionally add a title using the "main" input. 

Let's make a bar chart based on our enrollment dataset. We want the height of each bar to be equal to the number of students in that faculty, so we use "data.fac$students" as the height input.

```r
barplot(height = data.fac$students, names.arg = data.fac$faculty, main = "Bar Chart of Student Enrollment by Faculty")
```

Another optional input for the bar plot function is called "las". If we set "las" equal to 2, then the bar labels are printed vertically instead of horizontally. The "las" input is only ever left out or set equal to 2, other values don't do anything.

Let's make the same bar plot, but this time with vertical bar labels.

```r
barplot(height = data.fac$students, names.arg = data.fac$faculty, 
        las = 2, main = "Bar Chart of Student Enrollment by Faculty")
```

The "las" input is especially helpful when our labels have long names that take up more space than the width of 1 bar.

The next dataset we will work with is from a cereal called Chocolate Frosted Sugar Bombs (CFSB for short). Boxes of CFSB are advertized as containing 20 oz. of cereal. In reality, each box differs slightly from this target. This dataset is a record of actual weights for 100 boxes of CFSB cereal. This is a quantitative variable. The individuals are boxes of cereal.

Rather than enter the data directly in our code, we will tell R where to look for this dataset using the function called "read.csv". This function reads the data from your file and automatically creates a data frame in R to store it. The "read.csv" function has 1 main input called "file". The "file" input tells R the name of the file containing your dataset. In our case, the name of this file is "CFSB.csv". It is important that you upload the dataset to the same folder as this notebook in your Jupyter server.

The "read.csv" function has an optional input too, called "header". The "header" input tells R whether the first row of your dataset file is the name of a variable, or just another observation. We can set "header" equal to TRUE (in all capitals) if the first row of our file is the variable's name. If we set "header" equal to FALSE (also in all capitals), then the first row of our file is treated as just another observation. The default value of "header" in the "read.csv" function is TRUE, so if we don't include it then R treats it as TRUE. When analyzing data, you have to check the data file to see which of these options is appropriate.

Let's now read the CFSB dataset into R using the read.csv function. We will set "header" equal to TRUE (again, in all capitals), so that the first row of our data file is interpreted as the name of a variable. You can open this dataset in Jupyter to make sure this is the right choice. Alternatively, we could just not include "header" in the "read.csv" function, because its default value here is TRUE.

```r
data.cfsb = read.csv(file="CFSB.csv", header=TRUE)
print(data.cfsb)
```

Most of this dataset looks pretty much the same, so we could have gotten away with just printing out the first few lines. The "head" function does this for us.

```r
head(data.cfsb)
```

The "head" function formats our data a little differently, but it gives us the same information as the "print" function (just less of it). This makes checking a dataset much more manageable.

## Histograms

Now that we've read in our data, let's learn something about it. Our variable is quantitative, so a histogram is an appropriate plot. We make histograms in R using the "hist" function. The "hist" function has one main input called "x". The "x" input tells R the values that our variable takes. For us, this will be the "Weight" variable in our "data.cfsb" data frame. We can also include a title for our histogram using the usual optional input called "main".

Let's make our histogram now.

```r
hist(x = data.cfsb$Weight, main="Histogram of Cereal Box Weights")
```

This histogram looks okay, but the axis labels could be better. The "hist" function has another optional input called "xlab". The "xlab" input lets us change the label on the x-axis (the horizontal axis). Similarly, we can use the optional input called "ylab" to change the label on the y-axis (the vertical axis). Let's re-make this histogram with better axis labels.

```r
hist(x = data.cfsb$Weight, main="Histogram of Cereal Box Weights",
    xlab = "Weight (oz)", ylab = "Number of Boxes")
```

## Entering Data: Brain Size

- "...sizes of a people's brains are associated with their intelligence"

Our next dataset comes from a psychology study, where the researchers want to identify whether the size of a person's brain affects their intelligence. They measured some other things as well, but we are going to focus on just the size of a person's brain and their IQ. The first variable in our dataset is the person's IQ. The second variable is the person's brain size, measured in number of pixels it takes up on an MRI scan. These are both quantitative variables. The individuals in this study are people.

We read in this dataset in essentially the same way as our last dataset, using the "read.csv" function. The name of our data file is "Brain_Size.csv". Unlike the CFSB dataset, the "Brain_Size.csv" file has no variable names. This means we need to set the "header" input to FALSE (in all capitals). We can then add variable names once we have read in the data.

Let's start by reading the dataset into a data frame in R, and print out the first few observations to make sure it worked. We call this data frame "data.BrS".

```r
data.BrS = read.csv(file = "Brain_Size.csv", header = FALSE)
head(data.BrS)
```

The current variable names, "V1" and "V2", are not very helpful. Let's change these names to make them more informative. We do this in R using the "names" function. If we just run the "names" function on our data frame, it gives us back the current variable names.

```r
names(data.BrS)
```

This can be helpful if you have a data frame with lots of variables and you want to check what they all are. However, we want to change these names. This is done by setting the "names" function equal to a list of the names that we want. The first variable is IQ and the second variable is brain size. This means that we need to make a list with two entries, the first being "IQ" and the second being brain size, or just "Size". We then make the names of our data frame equal to this list.

```r
names(data.BrS) = c("IQ", "Size")
```

Remember that the "c" function is the glue that makes everything inside it into a single list.

Let's print out the first few observations to make sure that our variables now have the right names.

```r
head(data.BrS)
```

That's exactly what we want. Now we can move on to graphing the data.

## Scatterplots

"axis labels"

The brain size dataset has two variables. We can plot histograms for each of these variables individually, and this can tell us something about the dataset. We know that the "hist" function plots a histogram, and we can use "$" to access different variables in a data frame. We can also change the title and axis lables using the "main", "xlab" and "ylab" inputs.

```r
hist(x = data.BrS$IQ, main = "Histogram of IQ", xlab = "IQ", ylab = "Number of People")
hist(x = data.BrS$Size, main = "Histogram of Brain Size", xlab = "Size (Number of Pixels)", ylab = "Number of People")
```

Notice that the IQ variable is split into two groups (called a bimodal distribution). This is because the researchers selected people that had an IQ either above 129 or below 104. 

These histograms tell us a lot about IQ and brain size separately, but we want to understand their relationship. For that, we need a scatterplot. Each individual has a score for IQ and for brain size. A scatterplot plots one of these variables on the x-axis and the other on the y-axis. It then draws a point for each individual with their IQ and brain size.

In R, we make scatterplots using the "plot" function. The two main inputs for the "plot" function are "x" and "y". The "x" input tells R all the values for the x-axis variable. The "y" input tells R all the values for the y-axis variable. We can also optionally include the "main", "xlab" and "ylab" inputs to set the title, x-axis label and y-axis label respectively.

Let's make a scatterplot now with brain size on the x-axis and IQ on the y-axis.

```r
plot(x = data.BrS$Size, y = data.BrS$IQ, main = "Scatterplot of IQ vs Brain Size",
    xlab = "Brain Size (Number of Pixels)", ylab = "IQ")
```

Notice that our data are divided into two groups. This matches what we saw in the histogram of IQ.
